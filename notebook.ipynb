Tujuan utama dari analisis ini adalah untuk membantu rumah sakit dalam memfokuskan panggilan tindak lanjut dan perhatian pada pasien-pasien yang memiliki risiko tinggi untuk mengalami readmisi, sehingga memungkinkan penyedia layanan kesehatan untuk mengambil tindakan pencegahan yang sesuai dan meningkatkan kualitas perawatan pasien.

4. Implementasi dan Tindak Lanjut: Mengimplementasikan model prediktif ke dalam sistem informasi rumah sakit untuk mendukung pengambilan keputusan. Selain itu, merancang strategi tindak lanjut yang efektif berdasarkan hasil prediksi untuk mengurangi risiko readmisi pasien, termasuk pengaturan panggilan tindak lanjut, perubahan rencana perawatan, atau penyuluhan kepada pasien.

Dengan pendekatan ini, diharapkan permasalahan mengenai readmisi pasien dapat diselesaikan dengan lebih efektif, sehingga rumah sakit dapat memberikan perawatan yang lebih baik dan meminimalkan beban serta biaya yang terkait dengan readmisi pasien.

**5.2 Some doctors believe diabetes might play a central role in readmission. Explore the effect of a diabetes diagnosis on readmission rates.**

```python
plt.figure(figsize=(12,7))
sns.countplot(x='diag_1', hue='readmitted', data=df)
plt.title("Readmitted by Primary Diagnosis")
plt.xlabel(None)
plt.ylabel(None)

plt.show()
```

Berdasarkan jumlah pasien yang readmisi terhadap diagnosis utama, diperoleh bahwa **diabetes** merupakan satu-satunya diagnosis utama yang memiliki pengaruh signifikan terhadap readmisi pasien. Hal ini ditunjukkan dengan jumlah pasien yang mengalami readmisi lebih tinggi dibandingkan dengan jumlah pasien yang tidak mengalami readmisi.

```python
df_diabetes = df[df['diag_1'] == 'Diabetes']

plt.figure(figsize=(6,4))
ax = sns.countplot(x='readmitted', data=df_diabetes,)
ax.set_title("Readmitted by Diabetes Diagnosis")
ax.set_xlabel(None)
ax.set_ylabel(None)
ax.set_ylim(0,1050)
t = float(len(df_diabetes))
for p in ax.patches:
    height = p.get_height()
    ax.text(x=p.get_x() + p.get_width() / 2.,
            y=height + 10,
            s='{:.2f}%'.format((height / t) * 100),
            fontdict={'fontsize':10, 'fontweight':'bold'},
            ha='center')

plt.show()
```

Perhatikan bahwa pasien dengan diagnosis utama diabetes memiliki persentase readmisi lebih besar dibandingkan dengan yang tidak readmisi. Namun, apakah perbedaan ini signifikan, yaitu sekitar 6% hingga 7%? Untuk menjawab pertanyaan ini, akan dilakukan uji statistik untuk menguji apakah perbedaan tersebut signifikan sehingga dapat disimpulkan bahwa diabetes memiliki pengaruh terhadap readmisi pasien.

```python
# Uji proporsi
df_diabetes['readmitted'] = df_diabetes['readmitted'].apply(lambda x: 1 if x == 'yes' else 0)
count_readmitted = df_diabetes['readmitted'].sum()
count_not_readmitted = len(df_diabetes) - count_readmitted

nobs = np.array([count_readmitted + count_not_readmitted, count_readmitted])
count = np.array([count_readmitted, count_readmitted])

stat, pval = proportions_ztest(count, nobs)

print(f"Jumlah pasien diabetes yang readmisi: {count_readmitted}")
print(f"Jumlah pasien diabetes yang tidak readmisi: {count_not_readmitted} \n")
print(f"Statistik uji: {stat}")
print(f"Nilai p: {pval} \n")

alpha = 0.05
if pval < alpha:
    print("Perbedaan persentase readmisi signifikan. Diabetes memiliki pengaruh terhadap readmisi pasien.")
else:
    print("Perbedaan persentase readmisi tidak signifikan. Diabetes tidak memiliki pengaruh yang signifikan terhadap readmisi pasien.")
```

Berdasarkan hasil uji proporsi dengan alpha 0.05, terdapat cukup bukti untuk menolak hipotesis nol. Oleh karena itu, dapat disimpulkan bahwa diabetes memiliki pengaruh yang signifikan terhadap keputusan readmisi pasien.

Perhatikan, pada fitur-fitur numerik secara visual dicari pengaruhnya terhadap keputusan readmisi pasien, diperoleh:

- **Jumlah Prosedur Laboratorium dan Prosedur Lainnya:** Tidak ditemukan perbedaan signifikan dalam jumlah prosedur laboratorium dan prosedur lainnya antara pasien yang readmisi dan yang tidak.

- **Kunjungan Sebelumnya:** Pasien yang tidak pernah melakukan kunjungan rawat jalan, rawat inap, atau ke unit gawat darurat dalam tahun sebelum tinggal di rumah sakit cenderung memiliki kemungkinan lebih kecil untuk readmisi. Sebaliknya, semakin banyak kunjungan yang dilakukan, semakin besar kemungkinan readmisi, menunjukkan bahwa riwayat kunjungan memiliki pengaruh pada readmisi.

```python
results = pd.DataFrame(columns=['feature', 'statistic', 'pvalue', 'corr'])
alpha = 0.05

for col in cat_features:
    obs = []
    contingency_table = pd.crosstab(index=df['readmitted'], columns=df[col])

    rest = chi2_contingency(contingency_table)
    corr = 'yes' if rest.pvalue < alpha else 'no'

    results = pd.concat([results,
                       pd.DataFrame({'feature':[col], 'statistic':[rest.statistic],
                                      'pvalue':[rest.pvalue], 'corr':[corr]})], ignore_index=True)

for col in num_features:
    group1 = df[df['readmitted'] == 'yes'][col]
    group2 = df[df['readmitted'] == 'no'][col]
    rest = ttest_ind(group1, group2)

    results = pd.concat([results,
                       pd.DataFrame({'feature':[col], 'statistic':[rest.statistic],
                                      'pvalue':[rest.pvalue], 'corr':[corr]})], ignore_index=True)

results = results.sort_values(by='pvalue', ascending=True).reset_index(drop=True)
results
```

Perhatikan, berdasarkan uji statistik yang dilakukan untuk mengetahui pengaruh fitur numerik ataupun kategori terhadap keputusan readmisi pasien, diperoleh bahwa semua fitur memiliki hubungan signifikan dengan readmisi. Berikut beberapa poin penting:

When we have an unbalanced data set where there are more negatives than positives or vice versa, it is important to first balance the data in some way to give it a balanced weight. There are 3 strategies that are usually used

- sub-sample the more dominant class: use a random subset of negatives
- over-sample the unbalanced class: use the same positive sample multiple times
- create synthetic positive data

Usually, we would use the last two methods if we only had a few positive cases. Since we have several thousand positive cases, let's use the sub-sample approach. Here, we will create a balanced training dataset that has 50% positives and 50% negatives.

## **7. Modelling**

Dalam upaya untuk memprediksi readmisi pasien, penting untuk memilih model yang dapat menangani kompleksitas dan variasi data dengan baik. Berdasarkan analisis awal, kita menyadari bahwa hubungan antara berbagai fitur (seperti jumlah rawat inap, diagnosis, dan tes laboratorium) dengan keputusan readmisi tidak selalu linear. Oleh karena itu, pemilihan model non-linear menjadi pilihan yang tepat.

1. **Regresi Logistik**:
   - **Alasan**: Regresi logistik adalah model yang sederhana dan interpretatif, yang dapat digunakan sebagai baseline dalam masalah klasifikasi.
   - **Implementasi**: Model regresi logistik akan dilatih dengan fitur-fitur asli. Parameter seperti regularisasi dan solver akan disetel menggunakan pencarian grid atau random search untuk kinerja optimal.

2. **Random Forest**:
   - **Alasan**: Random Forest adalah model ensemble yang menggabungkan beberapa decision trees, menangkap hubungan non-linear dan interaksi antar fitur dengan baik.
   - **Implementasi**: Model ini akan dilatih dengan sejumlah trees yang besar, dan parameter seperti jumlah trees dan depth akan disetel untuk menghindari overfitting.

3. **XGBoost (Extreme Gradient Boosting)**:
   - **Alasan**: XGBoost adalah varian dari gradient boosting yang menawarkan pendekatan boosting yang kuat, meningkatkan kinerja dengan menggabungkan prediksi dari model-model sederhana.
   - **Implementasi**: Model ini akan dilatih dengan mengoptimalkan parameter boosting, learning rate, dan jumlah estimators melalui teknik cross-validation.

4. **Neural Networks (Multilayer Perceptrons)**:
   - **Alasan**: Neural networks dengan beberapa lapisan (hidden layers) dapat mempelajari hubungan non-linear yang kompleks dalam data.
   - **Implementasi**: MLP akan dilatih dengan berbagai arsitektur (jumlah neurons dan layers), menggunakan teknik regularisasi seperti dropout untuk menghindari overfitting.

```python
def evaluate_train(model, X_train, y_train):
  model_name = type(model).__name__
  df = pd.DataFrame()
  start_time = time.time()
  model.fit(X_train, y_train)
  end_time = time.time()

  training_time = end_time - start_time

  scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']
  cv_results = cross_validate(model, X_train, y_train, cv=5, scoring=scoring)

  # evaluation training
  df.loc['accuracy', model_name] = cv_results['test_accuracy'].mean()
  df.loc['precision', model_name] = cv_results['test_precision'].mean()
  df.loc['recall', model_name] = cv_results['test_recall'].mean()
  df.loc['f1_Score', model_name] = cv_results['test_f1'].mean()
  df.loc['roc_auc_score', model_name] = cv_results['test_roc_auc'].mean()
  df.loc['time (seconds)', model_name] = training_time

  return df

lr = LogisticRegression(max_iter=1000, solver='lbfgs', C=1.0)
rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
xgb = XGBClassifier(n_estimators=100, max_depth=3, learning_rate=0.1, random_state=42)
mlp = MLPClassifier(hidden_layer_sizes=(10,10), max_iter=1000, activation='relu', random_state=42)
```

```python
train_lr = evaluate_train(lr, X_train_tf, y_train_sm)
train_rf = evaluate_train(rf, X_train_tf, y_train_sm)
train_xgb = evaluate_train(xgb, X_train_tf, y_train_sm)
train_mlp = evaluate_train(mlp, X_train_tf, y_train_sm)
```

```python
result_train = pd.concat([train_lr, train_rf, train_xgb, train_mlp], axis=1)
result_train
```

```python
def evaluate_test(model, X_test, y_test):
  model_name = type(model).__name__
  df = pd.DataFrame()
  start_time = time.time()
  y_pred = model.predict(X_test)
  end_time = time.time()

  predicting_time = end_time - start_time

  # evaluation training
  df.loc['accuracy T', model_name] = accuracy_score(y_test, y_pred)
  df.loc['precision T', model_name] = precision_score(y_test, y_pred)
  df.loc['recall T', model_name] = recall_score(y_test, y_pred)
  df.loc['f1_score T', model_name] = f1_score(y_test, y_pred)

  if hasattr(model, "predict_proba"):
      df.loc['roc_auc_score T', model_name] = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
  else:
      df.loc['roc_auc_score T', model_name] = None

  df.loc['time T (seconds)', model_name] = predicting_time

  return df

X_test_tf = scaler.transform(X_test)
```

```python
test_lr = evaluate_test(lr, X_test_tf, y_test)
test_rf = evaluate_test(rf, X_test_tf, y_test)
test_xgb = evaluate_test(xgb, X_test_tf, y_test)
test_mlp = evaluate_test(mlp, X_test_tf, y_test)
```

```python
result_test = pd.concat([test_lr, test_rf, test_xgb, test_mlp], axis=1)
result_test
```

### Model Selection: Hyperparameter Tuning

```python
# Parameter distribution for SVM with RBF kernel
param_dist_svc = {
    'C': np.logspace(-3, 3, 7),  # Regularization parameter
    'gamma': np.logspace(-4, 1, 6),  # Kernel coefficient
}

# Parameter distribution for MLP
param_dist_mlp = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 100)],  # Number of neurons in the hidden layers
    'activation': ['relu', 'tanh'],  # Activation functions
    'solver': ['adam', 'sgd'],  # Optimization algorithms
    'alpha': np.logspace(-4, -1, 4),  # Regularization parameter
    'learning_rate': ['constant', 'adaptive'],  # Learning rate schedule
}

svc = SVC(kernel='rbf', probability=True)
mlp = MLPClassifier(max_iter=1000)
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import make_scorer, f1_score
```

```python
# Define scorers
scorer = make_scorer(f1_score)

# Randomized search for SVM
random_search_svc = RandomizedSearchCV(
    estimator=svc,
    param_distributions=param_dist_svc,
    n_iter=50,  # Number of parameter settings sampled
    scoring=scorer,
    cv=5,  # Number of folds in cross-validation
    verbose=2,
    n_jobs=-1,  # Use all available CPUs
    random_state=42
)

# Randomized search for MLP
random_search_mlp = RandomizedSearchCV(
    estimator=mlp,
    param_distributions=param_dist_mlp,
    n_iter=50,  # Number of parameter settings sampled
    scoring=scorer,
    cv=5,  # Number of folds in cross-validation
    verbose=2,
    n_jobs=-1,  # Use all available CPUs
    random_state=42
)
```
